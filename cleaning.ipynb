```python
import pandas as pd

viewed_posts = pd.read_csv("viewed_posts.csv")
print(viewed_posts.isnull().sum())  # Check for missing values
print(viewed_posts.head())
```

```python
summary_posts = pd.read_csv("summary_posts.csv")
# print(summary_posts.isnull().sum())  # Check for missing values
print(summary_posts.columns)
```

```python
print(len(summary_posts))
```

```python
threshold = 0.7

non_missing_percentage = summary_posts.notnull().mean()

columns_to_keep = non_missing_percentage[non_missing_percentage > threshold].index


filtered_df = summary_posts[columns_to_keep]


print("Original DataFrame:")
print(summary_posts.columns)

print("\nFiltered DataFrame:")
print(filtered_df.columns)
```

```python
pd.set_option('display.max_columns', None)


filtered_df.head()
```

```python
filtered_df.columns
```

```python
print(filtered_df.isnull().sum())  # Check for missing values
```

```python
filtered_df.drop(columns=['post_summary_other_person_details', 'post_summary_main_person_details_name','post_summary_main_person_details_gender'], inplace=True)
filtered_df.drop(columns=['post_summary_is_face_exist','post_summary_no_of_person_in_video'], inplace=True)
filtered_df.drop(columns=['category_image_url','picture_url','thumbnail_url','identifier','is_locked','first_name','last_name'], inplace=True)
filtered_df.drop(columns=['username'], inplace=True)
```

```python
filtered_df['post_summary_estimated_duration'].fillna(filtered_df['post_summary_estimated_duration'].mean, inplace=True)
```

```python
filtered_df.to_csv('cleaned_posts_summary.csv', index=False)
```
